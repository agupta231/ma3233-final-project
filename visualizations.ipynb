# Visualization Playground

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
from scipy import special as scs

%matplotlib notebook
```

## Runtime efficency comparision

```python
v = np.linspace(0, 20, 100).astype(int)
e = np.linspace(0, 20, 100).astype(int)
V, E = np.meshgrid(v, e)

naive = 2 ** E
quantum = np.sqrt(scs.comb(E, V))

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(V, E, quantum, cmap='hot')

ax.set_title('Grover\'s Algorithm Iterations')
ax.set_zscale('log')

ax.set_xlabel('V')
ax.set_ylabel('E')
ax.set_zlabel('Iterations')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(V, E, naive)
ax.plot_surface(V, E, quantum, cmap='hot')

ax.set_title('Grover\'s Algorithm vs. Naive Solution')
ax.set_zscale('log')

ax.set_xlabel('V')
ax.set_ylabel('E')
ax.set_zlabel('Iterations')
```
